```python
#i.
import numpy as np
gfg = np.matrix('[4,1,9;12,3,1;4,5,6]')
total_sum = 0
rows, cols = gfg.shape

for i in range(rows):
    for j in range(cols):
        total_sum += gfg[i, j]

print(total_sum)
```

```python
import numpy as np
array=np.array([10,52,62,16,16,54,453])
#i.
sorted=np.sort(array)
print(sorted)
print()

#ii.
l=list(array)
for i in sorted:
    if(i in l):
        print(l.index(i),end=' ')
print()

#iii.
for i in range(0,5):
    if(sorted[i]==sorted[i+1]):
        continue
    print(sorted[i])

n=len(sorted)
print()

#iv.
for i in range(1, 7):
    if n - i - 1 >= 0 and sorted[n - i] == sorted[n - i - 1]:
        continue
    print(sorted[n - i])
```

```python
import numpy as np
#a
ascii=ord('K')
print(ascii)
ascii2=ord('G')
asciii=ascii+ascii2
dataset=[]
dataset.append(asciii)
for i in range(1,5):
    dataset.append(asciii+50*i)
dataset=np.array(dataset)
print(dataset)

#b
tax=((asciii%5)+5)/100
print(tax)

tax_array=dataset*tax
print(tax_array)

#c
sales=asciii+100
discount=[]
for i in dataset:
    if(i==sales):
        discount.append(i-i*0.1)
print(discount)

sales_week=np.array([100,200,300,400,500,200,400,600,800,1000,400,800,1200,1600,2000])
sales_week = sales_week.reshape(3, 5)
print("the sales of the 3 weeks are :")
print(sales_week)
week_inc= np.array([1.02,1.04,1.06]).reshape(3,1)
updated=sales_week*week_inc
print("the updated sales are : ")
print(updated)
```
